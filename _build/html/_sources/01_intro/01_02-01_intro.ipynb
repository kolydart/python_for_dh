Κάθε φορά που γράφετε ένα πρόγραμμα ή κάνετε κάτι με κώδικα, ουσιαστικά γράφετε εντολές για φόρτωση, αποθήκευση, αλληλεπίδραση και χειρισμό δεδομένων με κάποιο τρόπο. Η κατανόηση των διαφορετικών τύπων δεδομένων και του τρόπου δόμησης και αποθήκευσής τους είναι, επομένως, απαραίτητη.
