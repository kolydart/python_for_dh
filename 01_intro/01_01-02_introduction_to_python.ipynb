# Εισαγωγή στην Python

## Γιατί Θα Έπρεπε οι Ανθρωπιστές να Μάθουν Προγραμματισμό;

Πριν ξεκινήσουμε αυτό το βιβλίο, ας ξεκινήσουμε με ένα απλό ερώτημα. *Γιατί θα έπρεπε οι ανθρωπιστές να μάθουν προγραμματισμό;* Για να το απαντήσουμε, ας εξετάσουμε ένα ιδανικό έργο ψηφιακών ανθρωπιστικών σπουδών.

Σε ένα ιδανικό έργο ψηφιακών ανθρωπιστικών σπουδών, ένας ανθρωπιστής (ή ομάδα ανθρωπιστών) θα αξιοποιεί την **γνώση του πεδίου του** ή το εμπειρικό του πεδίο. Ως ειδικός πεδίου, αυτός ο ηγέτης έργου, γνωστός καλύτερα ως κύριος ερευνητής (PI), θα μεταφέρει την ιδέα του στην ομάδα του, η οποία θα αποτελείται από άλλους στο πεδίο του και συνήθως από έναν τεχνικό ηγέτη. Αυτός ο τεχνικός ηγέτης θα χειριστεί (ανάλογα με τη χρηματοδότηση) όλες τις τεχνικές πτυχές του έργου (ανάπτυξη ιστοσελίδων, προγραμματισμό, διαχείριση δεδομένων κ.λπ.) ή θα ηγηθεί μιας ομάδας για να χειριστεί όλες τις τεχνικές πτυχές. Αυτό είναι το ιδανικό σενάριο. Κάθε άτομο στην ομάδα κάνει αυτό που κάνει καλύτερα.

Ένα τέτοιο έργο απαιτεί δύο πράγματα που είναι σε υψηλή ζήτηση από τους περισσότερους ερευνητές: χρηματοδότηση και χρόνο. Ο αυτάρκης ανθρωπιστής που μπορεί να αξιοποιήσει την γνώση του πεδίου του και ταυτόχρονα να λειτουργήσει με τεχνική ικανότητα σε ένα έργο μειώνει δραστικά και τα δύο αυτά προβλήματα με διάφορους τρόπους. Για να καταλάβουμε πώς, πρέπει πρώτα να καταλάβουμε πώς λαμβάνουν χρηματοδότηση τα έργα ψηφιακών ανθρωπιστικών σπουδών.

### To Χρονοδιάγραμμα για την Έναρξη ενός Μεγάλου Έργου Ψηφιακής Ιστορίας

Τα μεγάλα ψηφιακά έργα συχνά διαρκούν χρόνια για να ξεκινήσουν. Αυτά τα έργα συχνά ξεκινούν με έναν ερευνητή (ή ερευνητές) που επιθυμούν να εξερευνήσουν ένα ερώτημα ή να δημιουργήσουν μια ψηφιακή πλατφόρμα για μια πτυχή της έρευνάς τους. Αυτοί οι μελετητές θα απευθυνθούν σε άλλους στο πεδίο τους για να μετρήσουν το ενδιαφέρον. Αυτή η διαδικασία μπορεί να διαρκέσει αρκετούς μήνες ή ένα χρόνο, ειδικά αν οι ερευνητές περιμένουν να παρουσιάσουν την ιδέα τους στο επόμενο συνέδριο στο πεδίο τους.

Εάν υπάρχει ενδιαφέρον για ένα τέτοιο έργο, ο κύριος ερευνητής (PI) του έργου θα βρει έναν τεχνικό ειδικό που μπορεί να του πει αν το έργο είναι δυνατό και να λάβει εκτίμηση του κόστους του. Εάν ο PI δεν έχει τεχνικό ειδικό και δεν μπορεί να βρει έναν που να σχετίζεται με το πανεπιστήμιό του, αυτό μπορεί να είναι προκλητικό και χρονοβόρο. Όπως θα δούμε, η σχέση μεταξύ PI και τεχνικού ηγέτη είναι ένας μοναδικός δεσμός που απαιτεί καλή επικοινωνία και από τα δύο μέρη. Η επιλογή του σωστού τεχνικού ηγέτη είναι σημαντική γιατί η αμοιβή του μπορεί να είναι ένα από τα μεγαλύτερα στοιχεία του προϋπολογισμού. Είναι επίσης το άτομο που κάνει την όραση του PI ψηφιακή πραγματικότητα.

Στη συνέχεια, ο PI θα δημιουργήσει μια λίστα ευκαιριών χρηματοδότησης για το έργο και μετά θα αρχίσει να υποβάλλει αιτήσεις για αυτές. Κάθε αίτηση μπορεί να διαρκέσει αρκετούς μήνες για να ολοκληρωθεί. Μόλις ολοκληρωθεί, ο PI θα υποβάλει συνήθως την αίτηση στον φορέα χρηματοδότησης μέσω του πανεπιστημίου του. Εάν υπάρχουν πολλαπλοί PI στο έργο σε διαφορετικά πανεπιστήμια, αυτή η διαδικασία μπορεί να είναι πιο περίπλοκη καθώς όλα τα ιδρύματα συνήθως πρέπει να επικοινωνούν μεταξύ τους.

Το πανεπιστήμιο συνήθως εξετάζει όλα τα υλικά. Θα ενδιαφέρονται ιδιαίτερα για τον προϋπολογισμό. Αυτή η διαδικασία μπορεί να διαρκέσει μερικές εβδομάδες ενώ το πανεπιστήμιο διασφαλίζει ότι όλοι οι αριθμοί είναι σωστοί. Σε τελική ανάλυση, τα περισσότερα πανεπιστήμια θα λαμβάνουν ένα μέρος των χρημάτων της επιχορήγησης (μερικές φορές έως 30%), οπότε θέλουν να βεβαιωθούν ότι θα δουν τους αριθμούς πρώτα.

Μετά από αυτό, το πανεπιστήμιο θα υποβάλει την επιχορήγηση. Ανάλογα με τον φορέα χρηματοδότησης, ο χρόνος αναμονής μπορεί να κυμαίνεται από μερικούς μήνες έως περίπου ένα χρόνο. Κατά τη διάρκεια αυτού του χρόνου, το ψηφιακό έργο ενδέχεται να μην προχωρήσει πολύ ή καθόλου. Αντίθετα, μπορεί να παραμείνει σε περίοδο αναμονής ενώ ο PI περιμένει να ακούσει.

Όταν κοιτάζουμε αυτό το χρονοδιάγραμμα, σημαίνει ότι από την αρχική ιδέα έως την έναρξη του έργου, ακόμη και στο ιδανικό σενάριο, μπορεί να χρειαστούν ρεαλιστικά 2-3 χρόνια πριν ξεκινήσει το έργο. Όλα αυτά συνοδεύονται από έναν βαθμό κινδύνου. Ο PI πιστεύει στην ιδέα, έχει επίσης μετρήσει την κοινότητα για να γνωρίζει ότι υπάρχει ενδιαφέρον γι 'αυτήν, και έχει βρει έναν τεχνικό ηγέτη που μπορεί να κάνει δουλειά, αλλά πολλά πράγματα μπορούν να συμβούν σε 2-3 χρόνια. Σε εκείνο το χρόνο, νέες εξελίξεις στο πεδίο του ειδικού πεδίου μπορεί να έχουν αλλάξει την τροχιά του έργου (ή, στη χειρότερη περίπτωση, να το έχουν καταστήσει περιττό). Επιπλέον, νέες ψηφιακές προσεγγίσεις/μέθοδοι μπορεί να σημαίνουν ότι απαιτείται μια εντελώς νέα προσέγγιση. Αυτό είναι ιδιαίτερα συνηθισμένο σήμερα αν το έργο περιλαμβάνει μηχανικό μάθημα, ένα πεδίο στο οποίο νέες προόδους συμβαίνουν κάθε εβδομάδα. Επιπλέον, ο τεχνικός ηγέτης ενδέχεται να έχει πάρει δουλειά κάπου αλλού. (Αυτό συμβαίνει καθώς δεν συμβάλλονται έως ότου λάβουν την επιχορήγηση). Τέλος, δεν υπάρχει καμία εγγύηση ότι οι φορείς χρηματοδότησης εκτιμούν την ιδέα. Αυτό σημαίνει ότι ο χρόνος που δαπανήθηκε για το σχεδιασμό του έργου έννοιολογικά και την προετοιμασία για να ξεκινήσει θα μπορούσε να είναι εντελώς μάταιος.

Με τη μάθηση προγραμματισμού, συχνά μπορείτε να αρνηθείτε ολόκληρη αυτή τη διαδικασία δοκιμάζοντας την ιδέα σας εντός μερικών ημερών ή εβδομάδων. Αυτό σημαίνει ότι θα είστε σε θέση να ελέγξετε την εγκυρότητα της ιδέας και δυνητικά ακόμη και να δημιουργήσετε ένα **ελάχιστα βιώσιμο προϊόν (MVP)** που μπορεί να χρησιμοποιηθεί για να βοηθήσει στην απόκτηση χρηματοδότησης. Ένα MVP είναι ένα εργαζόμενο υποσύνολο ενός έργου για να αποδείξει την εγκυρότητά του. Για ένα έργο ψηφιακών ανθρωπιστικών σπουδών, αυτό μπορεί να σημαίνει μια απλή εφαρμογή προσκηνίου ή ένα υποσύνολο των δεδομένων που προετοιμάστηκε με έναν συγκεκριμένο τρόπο. Τα MVP είναι αρκετά κοινά σήμερα σε μέρη όπως το Silicon Valley, όπου χρησιμοποιούνται MVP για να παρουσιάσουν μια ιδέα σε επενδυτές. Αυτό βοηθά να δώσει μια πραγματική αίσθηση του τρόπου λειτουργίας του έργου. Ενώ δεν θα είναι διαθέσιμη όλη τη μηχανική σε ένα MVP, είναι μια καλή απόδειξη ιδέας. Αποδεικνύει επίσης ότι το έργο είναι δυνατό, η ομάδα μπορεί να κάνει τη δουλειά που υπόσχεται και ότι η ομάδα ήδη έχει μια καλή κατανόηση των δυνητικών ζητημάτων που ενδέχεται να προκύψουν. Αυτά είναι όλα τα πράγματα που αγαπούν οι φορείς χρηματοδότησης. Σε όλο το εγχειρίδιο αυτό, θα αναπτύξετε τις δεξιότητες που απαιτούνται για να δημιουργήσετε ένα MVP που μπορούν να παρουσιάσετε σε έναν φορέα χρηματοδότησης. Αυτό ισχύει ιδιαίτερα στο τελευταίο κεφάλαιο, όπου μαθαίνουμε πώς να δημιουργούμε εφαρμογές Python που βασίζονται στο νέφος με το Streamlit.


### Ο Αυτάρκης Ανθρωπιστής του Ψηφιακού Αιώνα

Σε τελική ανάλυση, η δυνατότητα προγραμματισμού μετατρέπει τον ανθρωπιστή σε έναν απόλυτα αυτάρκη ερευνητή. Με τη μάθηση προγραμματισμού, μπορείτε να κάνετε πολλά περισσότερα από ό,τι δοκιμή ιδεών και δημιουργία MVP. Μπορείτε να μειώσετε σημαντικά το κόστος ενός έργου ψηφιακών ανθρωπιστικών σπουδών. Ως ειδικός πεδίου, γνωρίζετε το πεδίο σας καλά. Ως κωδικοποιητής, θα γνωρίζετε τι είναι τεχνικά δυνατό. Εάν καταλήξετε να κερδίσετε ειδίκευση στην επιστήμη δεδομένων ή τη μηχανική μάθηση (κάτι που είναι αρκετά ρεαλιστικό σήμερα καθώς και τα δύο πεδία είναι πιο προσιτά από ποτέ), μπορείτε να αξιοποιήσετε ορισμένες από τις πιο ισχυρές μεθόδους σε τεχνικά πεδία σε ένα πλαίσιο ψηφιακών ανθρωπιστικών σπουδών. Μπορείτε επίσης να εξαλείψετε πολλά προβλήματα που ενδέχεται να προκύψουν κατά τη διάρκεια ενός έργου ψηφιακών ανθρωπιστικών σπουδών.

Ένα πρόβλημα είναι η επικοινωνία. Ο κύριος ερευνητής του έργου πρέπει να έχει τη δυνατότητα να μεταφέρει τις ιδέες του στον τεχνικό ηγέτη. Αυτή η επικοινωνία συχνά γίνεται σε μια υβριδική γλώσσα που αλλάζει από ομάδα σε ομάδα. Ο PI με την πάροδο του χρόνου μπορεί να αρχίσει να χρησιμοποιεί τεχνικούς όρους και ο τεχνικός ηγέτης θα αρχίσει να χρησιμοποιεί ορισμένους όρους συγκεκριμένους του πεδίου. Αυτή η σχέση και η επικοινωνία χρειάζονται χρόνο για να δημιουργηθούν. Με τη σύμπτυξη ισχυρής κυριαρχίας των τεχνικών πτυχών ενός έργου, ο PI μπορεί είτε να εξαλείψει τη χρειάζεται για έναν τεχνικό ηγέτη είτε να μεταφέρει τις ιδέες του στην τεχνική πλευρά του έργου σε τεχνική γλώσσα.

Ένα άλλο πρόβλημα που εξαφανίζεται είναι ο χρόνος υστέρησης από ιδέα έως εκτέλεση. Μόλις ξεκινήσει ένα χρηματοδοτούμενο έργο, ο PI θα καθοδηγήσει την όραση και θα μεταφέρει κάθε βήμα στον τεχνικό ηγέτη του οποίου η δουλειά είναι να κάνει την όραση πραγματικότητα. Υπάρχει μια υστέρηση που υπάρχει εδώ. Συχνά, συσκέψεις σε ψηφιακά έργα συμβαίνουν σε κανονική βάση, μερικές φορές εβδομαδιαία ή δεκαπενθήμερη. Κατά τη διάρκεια αυτού του χρόνου, ζητήματα ενδέχεται να προκύψουν που τοποθετούνται στο πλάι έως την επόμενη σύσκεψη. Αυτό μπορεί να βάλει ένα ψηφιακό έργο πίσω από το χρονοδιάγραμμα. Εάν, ωστόσο, ο ανθρωπιστής είναι αυτάρκης, μπορούν να αναπτύξουν την ιδέα τους ανεξάρτητα από έναν τεχνικό ηγέτη. Αυτό σημαίνει ότι όταν προκύψει ένα πρόβλημα, ο PI μπορεί να το διορθώσει αμέσως και να μεταφέρει τη λύση στον τεχνικό ηγέτη.

Τρίτον, οι τεχνικοί ηγέτες σε έργα συνήθως απορροφούν πολλά κεφάλαια. Η προσλαμβάνοντας ενός προγραμματιστή μπορεί να είναι δαπανηρή, αλλά εάν το έργο σας απαιτεί επιστήμη δεδομένων και μηχανική μάθηση, ειδικοί σε αυτούς τους τομείς μπορεί να κοστίσουν εκατοντάδες δολάρια την ώρα. Εάν ο PI έχει τεχνική ειδίκευση, μπορεί να εκτελέσει τεχνικές πτυχές ενός έργου έως ότου λάβει αρκετά χρήματα για να προσλάβει έναν τεχνικό ηγέτη.


### Άλλα Οφέλη

Η μάθηση προγραμματισμού δίνει στον ανθρωπιστή εργαλεία πέρα από τη δυνατότητα μόνο προγραμματισμού. Αλλάζει θεμελιωδώς τον τρόπο που ο ανθρωπιστής βλέπει το πεδίο του και τα ερωτήματα που μπορεί να εξερευνήσει. Πράγματα που φαίνονταν αδύνατα πριν, θα φαίνονται ξαφνικά αρκετά απλά να λυθούν.

Ένα από τα μεγαλύτερα πλεονεκτήματα της μάθησης προγραμματισμού είναι η **αυτοματοποίηση**, ή η διαδικασία με την οποία γράφουμε κανόνες για ένα λογιστικό σύστημα να εκτελέσει μια επαναλαμβανόμενη εργασία. Ως ανθρωπιστές, κάνουμε πολλά πράγματα επαναληπτικά. Πολλές από τις εργασίες που πρέπει να εκτελέσουμε ως ανθρωπιστές είναι επαναλαμβανόμενες και εμείς, ως άνθρωποι, είμαστε επιρρεπείς στο να κάνουμε λάθη. Η ικανότητα αυτοματοποίησης αυτών των εργασιών μπορεί να μειώσει δραστικά τον χρόνο που ξοδεύουμε στην εκτέλεση επαναλαμβανόμενων εργασιών, από ώρες, εβδομάδες ή ακόμη και χρόνια, σε δευτερόλεπτα, λεπτά ή ώρες. Φανταστείτε ότι πρέπει να λάβετε δεδομένα από ένα αρχείο ιστοσελίδας. Φανταστείτε ότι τα πληροφορίες που χρειάζεστε βρίσκονται σε 2.000 διαφορετικές σελίδες. Πόσο χρόνο θα χρειαστεί να πάτε σε κάθε μία από αυτές τις σελίδες και να αντιγράψετε και να επικολλήσετε το κείμενο σε ένα έγγραφο Word; Στην Python, η εργασία μπορεί να κωδικοποιηθεί σε λεπτά και να αφεθεί να εκτελεστεί για μια ώρα. Αυτό σας επιτρέπει, τον ερευνητή να πάτε και να κάνετε κάποια άλλη πιο απαραίτητη εργασία· ή, ίσως, να απολαύσετε ένα ωραίο τσάι σε μια αιώρα ώστε όταν επιστρέψετε για να αναλύσετε τα έγγραφα, θα ήσαστε ανακεφαλαιωμένοι.

Ο προγραμματισμός δεν μας επιτρέπει απλώς να αυτοματοποιούμε εργασίες όπως αυτή. Μας επιτρέπει επίσης να καθαρίζουμε συστηματικά δεδομένα. Φανταστείτε ότι θέλατε να αναζητήσετε διαθέσιμα σάρωση PDF της μεσαιωνικής λατινικής. Αυτό το πρόβλημα παρουσιάζει πολλά βασικά ζητήματα που καθιστούν μια τέτοια εργασία αδύνατη ή αναξιόπιστη. Πρώτον, η μεσαιωνική λατινική δεν είχε καμία σταθερή σύμβαση ορθογραφίας. Αυτό σημαίνει ότι ορισμένες σαρώσεις ενδέχεται να έχουν διαφορετικές ορθογραφίες λέξεων. Δεύτερον, τα λατινικά είναι μια ιδιαίτερα δεσμευτική γλώσσα, που σημαίνει ότι η σειρά των λέξεων δεν είναι σημαντική, παρά η λήξη μιας δεδομένης λέξης. Όταν συνδυάζεται με διαφορετικές ορθογραφίες, αυτό σημαίνει ότι κάθε λέξη μπορεί να αναπαρασταθεί μερικές φορές εκατοντάδες διαφορετικούς τρόπους. Επιπλέον, τα κείμενά σας είναι σαρώσεις. Εάν μετά, τότε οι σαρώσεις μπορεί να είναι σε τόσο κακή κατάσταση όπου το OCR δεν είναι ακριβές. Επιπλέον, οποιοδήποτε σύστημα OCR θα διατηρήσει αλλαγές γραμμής, που σημαίνει ότι εάν η λέξη-κλειδί που θέλετε να αναζητήσετε είναι τονισμένη επειδή η λέξη αναγνωρίζεται μεταξύ δύο γραμμών, πρέπει να λάβετε αυτό υπόψη στην αναζήτησή σας. Στη συνέχεια, πρέπει να λάβουμε υπόψη τις σημειώσεις του επεξεργαστή, οι οποίες συχνά σημειώνονται σε αγκύλες, παρενθέσεις και καρότα. Ενώ αυτό το παράδειγμα είναι σίγουρα ένα περίπλοκο, είναι τέλεια κοινό. Όλες οι γλώσσες προγραμματισμού δεν δημιουργούνται ίσες. Κάποιες είναι καλύτερα χρησιμοποιημένες για την ανάπτυξη λογισμικού, όπως η C· άλλες είναι καταλληλότερες για ανάπτυξη ιστοσελίδων, όπως τα HTML και JavaScript. Και άλλες είναι χαρακτηριστικά σε στατιστική ανάλυση, R (και Python). Που αριστεύει η Python; Η Python αριστεύει σε πολλές περιοχές. Ένα πράγμα που τη ξεχωρίζει από άλλες γλώσσες προγραμματισμού για νέους κωδικοποιητές είναι ότι είναι εύκολη στη χρήση και εύκολη στη γραφή.

Είναι σχετικά εύκολη να γράψει σε σχέση με άλλες γλώσσες προγραμματισμού επειδή η σύνταξη, ή ο τρόπος με τον οποίο γράφετε εργασίες στον κώδικα, είναι ευθύ. Είναι εύκολη στη ανάγνωση επειδή η Python χρησιμοποιεί αναγκαστική εσοχή. Αυτό σημαίνει ότι τα μπλοκ κώδικα που μπορούν να είναι δύσκολο να διαβαστούν σε άλλες γλώσσες προγραμματισμού, εντοπίζονται εύκολα στην Python.

Από τη δημιουργία του στις αρχές της δεκαετίας του 1990, η Python έχει αυξηθεί σημαντικά στη δημοτικότητα, η οποία, με τη σειρά της, έχει οδηγήσει σε μια μεγάλη κοινότητα προγραμματιστών και ένας μεγάλος αριθμός βιβλιοθηκών διατίθεται. Θα μάθουμε για τις βιβλιοθήκες αργότερα σε αυτό το εγχειρίδιο. Προς το παρόν, σκεφτείτε τις βιβλιοθήκες ως μεγάλες ποσότητες κώδικα που έχουν ήδη γραφτεί γι' ας ώστε να μπορείτε να γράψετε μία γραμμή κώδικα για να εκτελέσετε μια περίπλοκη εργασία που μπορεί να χρειαστεί εκατοντάδες ή χιλιάδες γραμμές κώδικα για να γραφεί.

Σήμερα, η Python είναι μια από τις πιο ευρέως χρησιμοποιούμενες γλώσσες προγραμματισμού και θεωρείται η απαραίτητη γλώσσα για ανάλυση κειμένου, μηχανική μάθηση, ανάλυση δεδομένων (παράλληλα με το R), διαδίκτυο scrapping, και πολλά άλλα.

Η Python (κατά το χρόνο της συγγραφής) είναι αυτήν τη στιγμή σε έκδοση `3.11.0`. Ας διαχωρίσουμε κάθε έναν από αυτούς τους αριθμούς. Το `3` αναφέρεται στην κύρια γλώσσα Python. Η Python 2 υπάρχει ακόμη (στην πραγματικότητα έρχεται ως πρότυπο σε όλα τα Mac), αλλά δεν υποστηρίζεται πλέον και αργά αντικαθίσταται από την Python 3. Δεν θα πρέπει να επενδύσετε χρόνο στη μάθηση της Python 2 καθώς χρησιμοποιείται μόνο για υποστήριξη legacy λογισμικού και έχει μεγάλη ποσότητα ζητημάτων ασφαλείας λόγω της καταργημένης κατάστασής της. Είναι σημαντικό να σημειώσουμε αυτό γιατί ορισμένα πράγματα κωδικοποιούνται διαφορετικά μεταξύ των δύο γλωσσών προγραμματισμού. Εάν έψαχνε βοήθεια σε ένα φόρουμ κωδικοποίησης, όπως StackOverflow, είναι σημαντικό να γνωρίζετε για αυτές τις διακρίσεις.

Το `11.0` στο `Python 3.11.0`, λέει μας ειδικά ποια έκδοση της Python 3. Δεν είναι ο κώδικας από ορισμένες βιβλιοθήκες συμβατός προς τα πίσω, που σημαίνει ότι ορισμένες βιβλιοθήκες απαιτούν συγκεκριμένη έκδοση της Python, οπότε η κατανόηση αυτής της έννοιας τώρα θα εξοικονομήσει σύγχυση αργότερα. Κάθε φορά που εμφανίζεται μια νέα έκδοση, νέα χαρακτηριστικά είναι διαθέσιμα, οπότε είναι σημαντικό να ενημερωθείτε με τρέχουσες εκδόσεις, αλλά δεν είναι πάντα απαραίτητο.
