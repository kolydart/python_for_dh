## Γιατί Θα Έπρεπε οι Ερευνητές των Ανθρωπιστικών Επιστημών να Μάθουν Προγραμματισμό;

Πριν ξεκινήσουμε αυτό το βιβλίο, ας ξεκινήσουμε με ένα απλό ερώτημα. *Γιατί θα έπρεπε οι επιστήμονες των ανθρωπιστικών επιστημών να μάθουν προγραμματισμό;* Για να το απαντήσουμε, ας εξετάσουμε ένα ιδανικό έργο ψηφιακών ανθρωπιστικών σπουδών.

Σε ένα ιδανικό έργο ψηφιακών ανθρωπιστικών σπουδών, ένας ερευνητής (ή ομάδα ερευνητών) θα αξιοποιεί την **γνώση του πεδίου του** ή το εμπειρικό του πεδίο. Ως ειδικός στο πεδίο του, αυτός ο υπεύθυνος του έργου, γνωστός καλύτερα ως κύριος ερευνητής (PI), θα μεταφέρει την ιδέα του στην ομάδα του, η οποία θα αποτελείται από άλλους στο πεδίο του και συνήθως από έναν τεχνικό ηγέτη. Αυτός ο τεχνικός ηγέτης θα χειριστεί (ανάλογα με τη χρηματοδότηση) όλες τις τεχνικές πτυχές του έργου (ανάπτυξη ιστοσελίδων, προγραμματισμό, διαχείριση δεδομένων κ.λπ.) ή θα ηγηθεί μιας ομάδας για να χειριστεί όλες τις τεχνικές πτυχές. Αυτό είναι το ιδανικό σενάριο. Κάθε άτομο στην ομάδα κάνει αυτό που κάνει καλύτερα.

Ένα τέτοιο έργο απαιτεί δύο πράγματα που είναι σε υψηλή ζήτηση από τους περισσότερους ερευνητές: χρηματοδότηση και χρόνο. Ο αυτάρκης ερευνητής που μπορεί να αξιοποιήσει τη γνώση του πεδίου του και ταυτόχρονα να λειτουργήσει με τεχνική ικανότητα σε ένα έργο μειώνει δραστικά και τα δύο αυτά προβλήματα με διάφορους τρόπους. Για να καταλάβουμε πώς, πρέπει πρώτα να καταλάβουμε πώς λαμβάνουν χρηματοδότηση τα έργα ψηφιακών ανθρωπιστικών σπουδών.

### To Χρονοδιάγραμμα για την Έναρξη ενός Μεγάλου Έργου Ψηφιακής Ιστορίας

Τα μεγάλα ψηφιακά έργα συχνά διαρκούν χρόνια για να ξεκινήσουν. Αυτά τα έργα συχνά ξεκινούν με έναν ερευνητή (ή ερευνητές) που επιθυμούν να εξερευνήσουν ένα ερώτημα ή να δημιουργήσουν μια ψηφιακή πλατφόρμα για μια πτυχή της έρευνάς τους. Αυτοί οι μελετητές θα απευθυνθούν σε άλλους στο πεδίο τους για να μετρήσουν το ενδιαφέρον. Αυτή η διαδικασία μπορεί να διαρκέσει αρκετούς μήνες ή ένα χρόνο, ειδικά αν οι ερευνητές περιμένουν να παρουσιάσουν την ιδέα τους στο επόμενο συνέδριο στο πεδίο τους.

Εάν υπάρχει ενδιαφέρον για ένα τέτοιο έργο, ο κύριος ερευνητής (PI) του έργου θα βρει έναν τεχνικό ειδικό που μπορεί να του πει αν το έργο είναι δυνατό και να λάβει εκτίμηση του κόστους του. Εάν ο PI δεν έχει τεχνικό ειδικό και δεν μπορεί να βρει έναν που να σχετίζεται με το πανεπιστήμιό του, αυτό μπορεί να είναι προκλητικό και χρονοβόρο. Όπως θα δούμε, η σχέση μεταξύ PI και τεχνικού ηγέτη είναι ένας μοναδικός δεσμός που απαιτεί καλή επικοινωνία και από τα δύο μέρη. Η επιλογή του σωστού τεχνικού ηγέτη είναι σημαντική γιατί η αμοιβή του μπορεί να είναι ένα από τα μεγαλύτερα στοιχεία του προϋπολογισμού. Είναι επίσης το άτομο που κάνει την όραση του PI ψηφιακή πραγματικότητα.

Στη συνέχεια, ο PI θα δημιουργήσει μια λίστα ευκαιριών χρηματοδότησης για το έργο και μετά θα αρχίσει να υποβάλλει αιτήσεις χρηματοδότησης. Κάθε αίτηση μπορεί να διαρκέσει αρκετούς μήνες για να ολοκληρωθεί. Μόλις ολοκληρωθεί, ο PI θα υποβάλει συνήθως την αίτηση στον φορέα χρηματοδότησης μέσω του πανεπιστημίου του. Εάν υπάρχουν πολλαπλοί PI στο έργο σε διαφορετικά πανεπιστήμια, αυτή η διαδικασία μπορεί να είναι πιο περίπλοκη καθώς όλα τα ιδρύματα συνήθως πρέπει να επικοινωνούν μεταξύ τους.

Το πανεπιστήμιο συνήθως εξετάζει όλα τα υλικά. Θα ενδιαφέρονται ιδιαίτερα για τον προϋπολογισμό. Αυτή η διαδικασία μπορεί να διαρκέσει μερικές εβδομάδες ενώ το πανεπιστήμιο διασφαλίζει ότι όλοι οι αριθμοί είναι σωστοί. Σε τελική ανάλυση, τα περισσότερα πανεπιστήμια θα λαμβάνουν ένα μέρος των χρημάτων της επιχορήγησης (μερικές φορές έως 30%), οπότε θέλουν να βεβαιωθούν ότι θα δουν τους αριθμούς πρώτα.

Μετά από αυτό, το πανεπιστήμιο θα υποβάλει την επιχορήγηση. Ανάλογα με τον φορέα χρηματοδότησης, ο χρόνος αναμονής μπορεί να κυμαίνεται από μερικούς μήνες έως περίπου ένα χρόνο. Κατά τη διάρκεια αυτού του χρόνου, το ψηφιακό έργο ενδέχεται να μην προχωρήσει πολύ ή καθόλου. Αντίθετα, μπορεί να παραμείνει σε περίοδο αναμονής ενώ ο PI περιμένει να ακούσει.

Όταν κοιτάζουμε αυτό το χρονοδιάγραμμα, σημαίνει ότι από την αρχική ιδέα έως την έναρξη του έργου, ακόμη και στο ιδανικό σενάριο, μπορεί να χρειαστούν ρεαλιστικά 2-3 χρόνια πριν ξεκινήσει το έργο. Όλα αυτά συνοδεύονται από έναν βαθμό κινδύνου. Ο PI πιστεύει στην ιδέα, έχει επίσης μετρήσει την κοινότητα για να γνωρίζει ότι υπάρχει ενδιαφέρον γι' αυτήν, και έχει βρει έναν τεχνικό ηγέτη που μπορεί να κάνει τη δουλειά, αλλά πολλά πράγματα μπορούν να συμβούν σε 2-3 χρόνια. Σε εκείνον τον χρόνο, νέες εξελίξεις στο πεδίο μπορεί να έχουν αλλάξει την τροχιά του έργου (ή, στη χειρότερη περίπτωση, να το έχουν καταστήσει περιττό). Επιπλέον, νέες ψηφιακές προσεγγίσεις/μέθοδοι μπορεί να σημαίνουν ότι απαιτείται μια εντελώς νέα προσέγγιση. Αυτό είναι ιδιαίτερα συνηθισμένο σήμερα αν το έργο περιλαμβάνει μηχανική μάθηση, ένα πεδίο στο οποίο νέες εξελίξεις συμβαίνουν κάθε εβδομάδα. Επιπλέον, ο τεχνικός ηγέτης ενδέχεται να έχει βρει δουλειά κάπου αλλού. (Αυτό συμβαίνει καθώς δεν συμβάλλονται έως ότου λάβουν την επιχορήγηση). Τέλος, δεν υπάρχει καμία εγγύηση ότι οι φορείς χρηματοδότησης εκτιμούν την ιδέα. Αυτό σημαίνει ότι ο χρόνος που δαπανήθηκε για το σχεδιασμό του έργου και την προετοιμασία για να ξεκινήσει θα μπορούσε να είναι εντελώς μάταιος.

Με τη μάθηση προγραμματισμού, συχνά μπορείτε να παρακάμψετε ολόκληρη αυτή τη διαδικασία δοκιμάζοντας την ιδέα σας εντός μερικών ημερών ή εβδομάδων. Αυτό σημαίνει ότι θα είστε σε θέση να ελέγξετε την εγκυρότητα της ιδέας και δυνητικά ακόμη και να δημιουργήσετε ένα **ελάχιστα βιώσιμο προϊόν (MVP)** που μπορεί να χρησιμοποιηθεί για να βοηθήσει στην απόκτηση χρηματοδότησης. Ένα MVP είναι ένα λειτουργικό υποσύνολο ενός έργου για να αποδείξει την εγκυρότητά του. Για ένα έργο ψηφιακών ανθρωπιστικών σπουδών, αυτό μπορεί να σημαίνει μια απλή εφαρμογή προσκηνίου ή ένα υποσύνολο των δεδομένων που προετοιμάστηκε με έναν συγκεκριμένο τρόπο. Τα MVP είναι αρκετά κοινά σήμερα σε μέρη όπως το Silicon Valley, όπου χρησιμοποιούνται MVP για να παρουσιάσουν μια ιδέα σε επενδυτές. Αυτό βοηθά να δώσει μια πραγματική αίσθηση του τρόπου λειτουργίας του έργου. Ενώ δεν θα είναι διαθέσιμη όλη η μηχανική σε ένα MVP, είναι μια καλή απόδειξη ιδέας. Αποδεικνύει επίσης ότι το έργο είναι δυνατό, η ομάδα μπορεί να κάνει τη δουλειά που υπόσχεται και ότι η ομάδα έχει ήδη μια καλή κατανόηση των δυνητικών ζητημάτων που ενδέχεται να προκύψουν. Αυτά είναι όλα τα πράγματα που αγαπούν οι φορείς χρηματοδότησης. Σε όλο το εγχειρίδιο αυτό, θα αναπτύξετε τις δεξιότητες που απαιτούνται για να δημιουργήσετε ένα MVP που μπορείτε να παρουσιάσετε σε έναν φορέα χρηματοδότησης. Αυτό ισχύει ιδιαίτερα στο τελευταίο κεφάλαιο, όπου μαθαίνουμε πώς να δημιουργούμε εφαρμογές Python που βασίζονται στο νέφος με το Streamlit.


### Ο Αυτάρκης Ερευνητής των Ψηφιακών Ανθρωπιστικών Επιστημών

Σε τελική ανάλυση, η δυνατότητα προγραμματισμού μετατρέπει τον ερευνητή των ανθρωπιστικών επιστημών σε έναν απόλυτα αυτάρκη επιστήμονα. Με τη μάθηση προγραμματισμού, μπορείτε να κάνετε πολλά περισσότερα από ό,τι δοκιμή ιδεών και δημιουργία MVP. Μπορείτε να μειώσετε σημαντικά το κόστος ενός έργου ψηφιακών ανθρωπιστικών σπουδών. Ως ειδικός στο πεδίο σας, γνωρίζετε το πεδίο σας καλά. Ως κωδικοποιητής, θα γνωρίζετε τι είναι τεχνικά δυνατό. Εάν καταλήξετε να κερδίσετε ειδίκευση στην επιστήμη δεδομένων ή τη μηχανική μάθηση (κάτι που είναι αρκετά ρεαλιστικό σήμερα καθώς και τα δύο πεδία είναι πιο προσιτά από ποτέ), μπορείτε να αξιοποιήσετε ορισμένες από τις πιο ισχυρές μεθόδους σε τεχνικά πεδία σε ένα πλαίσιο ψηφιακών ανθρωπιστικών σπουδών. Μπορείτε επίσης να εξαλείψετε πολλά προβλήματα που ενδέχεται να προκύψουν κατά τη διάρκεια ενός έργου ψηφιακών ανθρωπιστικών σπουδών.

Ένα πρόβλημα είναι η επικοινωνία. Ο κύριος ερευνητής του έργου πρέπει να έχει τη δυνατότητα να μεταφέρει τις ιδέες του στον τεχνικό ηγέτη. Αυτή η επικοινωνία συχνά γίνεται σε μια υβριδική γλώσσα που αλλάζει από ομάδα σε ομάδα. Ο PI με την πάροδο του χρόνου μπορεί να αρχίσει να χρησιμοποιεί τεχνικούς όρους και ο τεχνικός ηγέτης θα αρχίσει να χρησιμοποιεί ορισμένους όρους συγκεκριμένους του πεδίου. Αυτή η σχέση και η επικοινωνία χρειάζονται χρόνο για να δημιουργηθούν. Με την απόκτηση ισχυρής κυριαρχίας των τεχνικών πτυχών ενός έργου, ο PI μπορεί είτε να εξαλείψει τη χρειάζεται για έναν τεχνικό ηγέτη είτε να μεταφέρει τις ιδέες του στην τεχνική πλευρά του έργου σε τεχνική γλώσσα.

Ένα άλλο πρόβλημα που εξαφανίζεται είναι ο χρόνος υστέρησης από ιδέα έως εκτέλεση. Μόλις ξεκινήσει ένα χρηματοδοτούμενο έργο, ο PI θα καθοδηγήσει την όραση και θα μεταφέρει κάθε βήμα στον τεχνικό ηγέτη του οποίου η δουλειά είναι να κάνει την όραση πραγματικότητα. Υπάρχει μια υστέρηση που υπάρχει εδώ. Συχνά, συσκέψεις σε ψηφιακά έργα συμβαίνουν σε τακτική βάση, μερικές φορές εβδομαδιαία ή δεκαπενθήμερη. Κατά τη διάρκεια αυτού του χρόνου, ζητήματα ενδέχεται να προκύψουν που τοποθετούνται στο πλάι έως την επόμενη σύσκεψη. Αυτό μπορεί να καθυστερήσει ένα ψηφιακό έργο. Εάν, ωστόσο, ο ερευνητής είναι αυτάρκης, μπορεί να αναπτύξει την ιδέα του ανεξάρτητα από έναν τεχνικό ηγέτη. Αυτό σημαίνει ότι όταν προκύψει ένα πρόβλημα, ο PI μπορεί να το διορθώσει αμέσως και να μεταφέρει τη λύση στον τεχνικό ηγέτη.

Τρίτον, οι τεχνικοί ηγέτες σε έργα συνήθως απορροφούν πολλά κεφάλαια. Η πρόσληψη ενός προγραμματιστή μπορεί να είναι δαπανηρή, αλλά εάν το έργο σας απαιτεί επιστήμη δεδομένων και μηχανική μάθηση, ειδικοί σε αυτούς τους τομείς μπορεί να κοστίσουν εκατοντάδες δολάρια την ώρα. Εάν ο PI έχει τεχνική ειδίκευση, μπορεί να εκτελέσει τεχνικές πτυχές ενός έργου έως ότου λάβει αρκετά χρήματα για να προσλάβει έναν τεχνικό ηγέτη.

## Τι είναι η Python;

Η Python είναι μια γλώσσα προγραμματισμού. Οι γλώσσες προγραμματισμού είναι τρόποι που εμείς, ως άνθρωποι, μπορούμε να γράψουμε εντολές που στη συνέχεια θα εκτελεστούν από έναν υπολογιστή. Υπάρχουν πολλές διαφορετικές γλώσσες προγραμματισμού διαθέσιμες για ερευνητές να επιλέξουν από:

1. C
2. Python
3. JavaScript
4. R
5. HTML (αυτό είναι αμφιλεγόμενο)

Όλες οι γλώσσες προγραμματισμού δεν δημιουργούνται ίσες. Κάποιες είναι καλύτερα χρησιμοποιημένες για την ανάπτυξη λογισμικού, όπως η C· άλλες είναι καταλληλότερες για ανάπτυξη ιστοσελίδων, όπως τα HTML και JavaScript. Και άλλες είναι χαρακτηριστικά σε στατιστική ανάλυση, όπως η R (και η Python). Που αριστεύει η Python; Η Python αριστεύει σε πολλές περιοχές. Ένα πράγμα που τη ξεχωρίζει από άλλες γλώσσες προγραμματισμού για νέους κωδικοποιητές είναι ότι είναι εύκολη στη χρήση και εύκολη στη γραφή.

Είναι σχετικά εύκολη να γράψει σε σχέση με άλλες γλώσσες προγραμματισμού επειδή η σύνταξη, ή ο τρόπος με τον οποίο γράφετε εντολές στον κώδικα, είναι ευθύς. Είναι εύκολη στη ανάγνωση επειδή η Python χρησιμοποιεί αναγκαστική εσοχή. Αυτό σημαίνει ότι τα μπλοκ κώδικα που μπορούν να είναι δύσκολο να διαβαστούν σε άλλες γλώσσες προγραμματισμού, εντοπίζονται εύκολα στην Python.

Από τη δημιουργία της στις αρχές της δεκαετίας του 1990, η Python έχει αυξηθεί σημαντικά στη δημοτικότητα, η οποία, με τη σειρά της, έχει οδηγήσει σε μια μεγάλη κοινότητα προγραμματιστών και ένα μεγάλο αριθμό βιβλιοθηκών που διατίθενται. Θα μάθουμε για τις βιβλιοθήκες αργότερα σε αυτό το εγχειρίδιο. Προς το παρόν, σκεφτείτε τις βιβλιοθήκες ως μεγάλες ποσότητες κώδικα που έχουν ήδη γραφτεί ώστε να μπορείτε να γράψετε μία γραμμή κώδικα για να εκτελέσετε μια περίπλοκη εργασία που μπορεί να χρειαστεί εκατοντάδες ή χιλιάδες γραμμές κώδικα για να γραφεί.

Σήμερα, η Python είναι μια από τις πιο ευρέως χρησιμοποιούμενες γλώσσες προγραμματισμού και θεωρείται η απαραίτητη γλώσσα για ανάλυση κειμένου, μηχανική μάθηση, ανάλυση δεδομένων (παράλληλα με το R), διαδίκτυο scraping, και πολλά άλλα.

Η Python (κατά το χρόνο της συγγραφής) είναι αυτήν τη στιγμή σε έκδοση `3.11.0`. Ας διαχωρίσουμε κάθε έναν από αυτούς τους αριθμούς. Το `3` αναφέρεται στην κύρια γλώσσα Python. Η Python 2 υπάρχει ακόμη (στην πραγματικότητα έρχεται ως προεπιλογή σε όλα τα Mac), αλλά δεν υποστηρίζεται πλέον και σταδιακά αντικαθίσταται από την Python 3. Δεν θα πρέπει να επενδύσετε χρόνο στη μάθηση της Python 2 καθώς χρησιμοποιείται μόνο για υποστήριξη legacy λογισμικού και έχει μεγάλη ποσότητα ζητημάτων ασφαλείας λόγω της καταργημένης κατάστασής της. Είναι σημαντικό να σημειώσουμε αυτό γιατί ορισμένα πράγματα κωδικοποιούνται διαφορετικά μεταξύ των δύο γλωσσών προγραμματισμού. Εάν ψάχνετε βοήθεια σε ένα φόρουμ κωδικοποίησης, όπως το StackOverflow, είναι σημαντικό να γνωρίζετε για αυτές τις διακρίσεις.

Το `11.0` στο `Python 3.11.0`, μας λέει ειδικά ποια έκδοση της Python 3 χρησιμοποιούμε. Ο κώδικας από ορισμένες βιβλιοθήκες δεν είναι συμβατός προς τα πίσω, που σημαίνει ότι ορισμένες βιβλιοθήκες απαιτούν συγκεκριμένη έκδοση της Python, οπότε η κατανόηση αυτής της έννοιας τώρα θα εξοικονομήσει σύγχυση αργότερα. Κάθε φορά που εμφανίζεται μια νέα έκδοση, νέα χαρακτηριστικά είναι διαθέσιμα, οπότε είναι σημαντικό να ενημερώνεστε με τρέχουσες εκδόσεις, αλλά δεν είναι πάντα απαραίτητο.

## Γιατί Python;

Για όλους τους παραπάνω λόγους, ενθαρρύνω τους ερευνητές των ανθρωπιστικών επιστημών να μάθουν την Python ως πρώτη γλώσσα προγραμματισμού τους. Είναι μια από τις ευκολότερες γλώσσες να μάθει κανείς, ευθεία να γράψει, και μπορεί να λύσει τα περισσότερα από τα προγραμματιστικά προβλήματα που ένας ερευνητής ενδέχεται να αντιμετωπίσει. Οι μεγάλες ποσότητες βιβλιοθηκών και οδηγών σημαίνουν ότι υπάρχουν λίγες εργασίες που θα αποδειχθούν αδύνατες να γίνουν.

Εάν έχετε πειστεί για την Python, τότε συνεχίστε να διαβάζετε αυτό το εγχειρίδιο καθώς μαθαίνουμε πώς να το εγκαταστήσουμε και να το χρησιμοποιήσουμε ως ερευνητές στις ανθρωπιστικές επιστήμες.
