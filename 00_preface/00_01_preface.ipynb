# Πρόλογος

Σχεδίασα αυτό το διδακτικό εγχειρίδιο για να εξυπηρετεί δύο λειτουργίες. Πρώτον, θα λειτουργήσει ως εισαγωγικό εγχειρίδιο στην Python για ανθρωπιστές (ή, ευρύτερα, για όσους δεν έχουν εμπειρία στον προγραμματισμό ή υπόβαθρο στην επιστήμη των υπολογιστών). Υπό αυτή την έννοια, οι αναγνώστες θα αποκτήσουν βασική κατανόηση των απαραίτητων πληροφοριών υποβάθρου, όπως τα δεδομένα και οι δομές δεδομένων (data structures), καθώς και των βασικών αρχών της Python.

Δεύτερον, το διδακτικό εγχειρίδιο έχει σχεδιαστεί όχι μόνο για να παρέχει στον αναγνώστη βασική κατανόηση της Python και του τρόπου χρήσης της, αλλά και του τρόπου εφαρμογής της ειδικά σε προβλήματα που αφορούν τις ανθρωπιστικές επιστήμες. Το βιβλίο εξερευνά ιδιαίτερα την εφαρμογή της Python στην ανάλυση δεδομένων (data analysis) με το Pandas, στην επεξεργασία φυσικής γλώσσας (natural language processing, NLP) με το spaCy, στη θεματική μοντελοποίηση (topic modeling) με το Gensim (για θεματική μοντελοποίηση LDA) και το Top2Vec (για σύγχρονη θεματική μοντελοποίηση), καθώς και στην ανάλυση κοινωνικών δικτύων (social network analysis) με τα NetworkX, Matplotlib και PyVis. Επιπλέον, διδάσκει τους αναγνώστες πώς να σχεδιάζουν γρήγορες εφαρμογές με το Streamlit και να τις αναπτύσσουν στο cloud μέσω του Streamlit Share.

Όλος ο κώδικας που παρουσιάζεται στο διδακτικό εγχειρίδιο είναι σχεδιασμένος να είναι όσο το δυνατόν πιο αναπαραγώγιμος. Στόχος είναι ο αναγνώστης να χρειάζεται μόνο να αντικαταστήσει τα προεπιλεγμένα δεδομένα με τα δικά του δεδομένα (ή με ελάχιστη προσπάθεια να μετατρέψει τα δεδομένα του σε δομημένη μορφή) για να επιτύχει παρόμοια αποτελέσματα. Έτσι, ο αναγνώστης θα μπορεί να εφαρμόζει τις μεθόδους που συζητούνται σε ολόκληρο το διδακτικό εγχειρίδιο με σχετική ευκολία στους δικούς του τομείς εξειδίκευσης.

Το Μέρος Ένα του διδακτικού εγχειριδίου εισάγει τον αναγνώστη στα βασικά της Python. Εδώ, θα μάθουμε τα θεμελιώδη στοιχεία του προγραμματισμού (Κεφάλαιο 1) και τα ουσιώδη για τα δεδομένα και τις δομές δεδομένων και πώς να εργαζόμαστε με αυτά μέσω της Python (Κεφάλαιο 2). Τα κεφάλαια αυτά θα παρέχουν την απαραίτητη βάση για την εξερεύνηση βασικών εννοιών του προγραμματισμού, όπως οι βρόχοι (loops) και οι δηλώσεις συνθήκης (conditionals) (Κεφάλαιο 3), οι συναρτήσεις (functions), οι κλάσεις (classes) και οι βιβλιοθήκες (libraries) (Κεφάλαιο 4), καθώς και την εργασία με εξωτερικά δεδομένα, όπως αρχεία κειμένου και JSON (Κεφάλαιο 5). Το τελευταίο κεφάλαιο του Μέρους Ένα θα εισαγάγει τον αναγνώστη στα βασικά του web scraping και της εργασίας με δεδομένα που βρίσκονται στο διαδίκτυο.

Μετά το Μέρος Ένα, ο αναγνώστης θα έχει αποκτήσει βασική κατανόηση της Python και της σύνταξής της, και θα μπορεί να αρχίσει να εργάζεται με δεδομένα για να σχεδιάσει έργα. Το υπόλοιπο του διδακτικού εγχειριδίου έχει σχεδιαστεί για να ενισχύσει όλες τις δεξιότητες που αποκτήθηκαν στο Μέρος Ένα. Κάθε ένα από τα επόμενα μέρη του διδακτικού εγχειριδίου θα εισαγάγει επίσης τον αναγνώστη στις βασικές βιβλιοθήκες που σχετίζονται με τα αντίστοιχα θέματά τους.

Στο Μέρος Δύο, θα πραγματοποιήσουμε εμβάθυνση στην ανάλυση δεδομένων. Στην Python, η βασική βιβλιοθήκη για την εργασία με δεδομένα, ειδικά δεδομένα σε πινακοποιημένη μορφή (tabular data), είναι το Pandas. Θα μάθουμε τα βασικά του Pandas εργαζόμενοι με το σύνολο δεδομένων ανοιχτού κώδικα του Τιτανικού. Μέχρι το τέλος του Μέρους Δύο, ο αναγνώστης θα έχει αποκτήσει κατανόηση του Pandas και θα μπορεί να το αξιοποιήσει στα δικά του έργα.

Το Μέρος Τρία μετατοπίζει την εστίαση στην ανάλυση κειμένου. Εδώ, θα μάθουμε για την επεξεργασία φυσικής γλώσσας (NLP) και πώς να χρησιμοποιούμε την Python και τη βιβλιοθήκη spaCy για να ασχοληθούμε με την NLP. Το συγκεκριμένο μέρος του διδακτικού εγχειριδίου δεν προϋποθέτει καμία προηγούμενη γνώση εκ μέρους του αναγνώστη σχετικά με την NLP ή τη γλωσσολογία. Ως εκ τούτου, θα παρέχει όλες τις βασικές πληροφορίες που απαιτούνται για να ξεκινήσουμε να εργαζόμαστε με κείμενα με πιο ισχυρούς τρόπους. Ο αναγνώστης θα μάθει για δύο διαφορετικές προσεγγίσεις στην NLP: τις βασισμένες σε κανόνες (heuristics) και τις βασισμένες σε μηχανική μάθηση. Και οι δύο εξυπηρετούν διαφορετικές λειτουργίες και πρέπει να χρησιμοποιούνται σε διαφορετικές περιπτώσεις. Μέχρι το τέλος αυτού του μέρους, ο αναγνώστης θα έχει αποκτήσει βασική κατανόηση της κάθε προσέγγισης και θα γνωρίζει πότε να τις χρησιμοποιεί. Ενώ οι προσεγγίσεις βασισμένες σε μηχανική μάθηση θα στηρίζονται στη χρήση έτοιμων μοντέλων spaCy, ο αναγνώστης θα μάθει πώς να χρησιμοποιεί κανόνες NLP για να δημιουργήσει προσαρμοσμένες λύσεις. Το τελευταίο κεφάλαιο αυτού του μέρους του διδακτικού εγχειριδίου θα εξετάσει ένα πρόβλημα του πραγματικού κόσμου: τη δημιουργία ενός ευρετικού pipeline βασισμένου σε κανόνες για την αναγνώριση και εξαγωγή συγκεκριμένων τύπων οντοτήτων από κείμενα.

Το Μέρος Τέσσερα μετατοπίζεται σε άλλες εφαρμογές της Python σε προβλήματα των ανθρωπιστικών επιστημών. Στο Κεφάλαιο 1 αυτού του μέρους, θα μάθουμε πώς να πραγματοποιούμε θεματική μοντελοποίηση, συγκεκριμένα θεματική μοντελοποίηση Latent Dirichlet Allocation (LDA), ώστε να αποκτήσουμε κατανόηση των βασικών εννοιών και της ιστορίας του πεδίου. Στη συνέχεια, θα μάθουμε για πιο πρόσφατες προσεγγίσεις στη θεματική μοντελοποίηση χρησιμοποιώντας μηχανική μάθηση με τη βιβλιοθήκη Top2Vec. Το Κεφάλαιο 2 θα εξετάσει την εκτέλεση ανάλυσης κειμένου σε μεγαλύτερα έγγραφα με το BookNLP. Στο Κεφάλαιο 3, θα εξετάσουμε την Ανάλυση Κοινωνικών Δικτύων με τα NetworkX και Matplotlib για την παραγωγή στατικών χαρτών. Θα μάθουμε επίσης πώς να δημιουργούμε δυναμικούς χάρτες δικτύου JavaScript και HTML με τη βιβλιοθήκη Python PyVis. Τα κεφάλαια αυτά έχουν σχεδιαστεί για να παρέχουν τις απαραίτητες γνώσεις υποβάθρου, την ορολογία και τον κώδικα Python που θα σας επιτρέψουν να αρχίσετε να εφαρμόζετε αυτές τις βιβλιοθήκες και μεθόδους στο δικό σας σύνολο δεδομένων.

Το Μέρος Πέντε του διδακτικού εγχειριδίου θα εισαγάγει τον αναγνώστη στην ανάπτυξη εφαρμογών (app development) με τη βιβλιοθήκη Streamlit. Οι αναγνώστες θα αποκτήσουν κατανόηση των βασικών του Streamlit και του τρόπου με τον οποίο μπορούν να αξιοποιήσουν τα στοιχεία του για να δημιουργήσουν προσαρμοσμένες εφαρμογές μέσα σε λίγες μόνο ώρες, οι οποίες μπορούν να φιλοξενηθούν στο cloud. Ο σκοπός αυτού του μέρους είναι να βοηθήσει τον αναγνώστη να μετατρέψει μια ιδέα από την εννοιολογική φάση στην πραγματικότητα στον συντομότερο δυνατό χρόνο.

Αφού ολοκληρώσετε αυτό το διδακτικό εγχειρίδιο, θα έχετε αποκτήσει αρκετά ισχυρή γνώση της Python ώστε να μπορείτε να την αξιοποιείτε στα δικά σας έργα. Θα έχετε επίσης ευρεία έκθεση σε διαφορετικούς τρόπους με τους οποίους η Python μπορεί να εφαρμοστεί σε προβλήματα των ανθρωπιστικών επιστημών. Τέλος, θα διαθέτετε τους απαραίτητους πόρους για να συνεχίσετε την εκπαίδευσή σας.

## Περιορισμοί αυτού του Διδακτικού Εγχειριδίου

Αν και αυτό το βιβλίο παρέχει επισκόπηση της Python, δεν καλύπτει όλες τις πτυχές της γλώσσας ή του τρόπου χρήσης της. Το βιβλίο έχει σχεδιαστεί για να σας εξοικειώσει με την Python όσο το δυνατόν ταχύτερα, παρέχοντάς σας τα απαραίτητα εργαλεία που χρειάζεστε για να διαβάζετε και να γράφετε στη γλώσσα προκειμένου να επιλύετε εργασίες γρήγορα και αποτελεσματικά. Το διδακτικό εγχειρίδιο δεν απευθύνεται σε επιστήμονες υπολογιστών που επιθυμούν να εξερευνήσουν σε βάθος τη γλώσσα προγραμματισμού, αλλά σε ανθρωπιστές που χρειάζονται την Python για να αυτοματοποιήσουν συγκεκριμένες εργασίες στη ροή εργασίας τους. Η εξήγηση ορισμένων πτυχών της γλώσσας περιορίζεται, συνεπώς, στο ελάχιστο.

Είναι σημαντικό να σημειωθεί ότι το βιβλίο αυτό είναι εξ ολοκλήρου βασισμένο σε Jupyter Notebooks (που παρουσιάζονται στο Μέρος Ένα, Κεφάλαιο Ένα). Αυτό σημαίνει ότι δεν θα εκτεθείτε στη γραμμή εντολών (command line) ούτε θα λάβετε κατάλληλη εκπαίδευση στη σύνταξη αρχείου Python script (.py). Αυτές είναι χρήσιμες δεξιότητες, αλλά όχι απαραίτητες για να ξεκινήσετε να εργάζεστε με δεδομένα. Παρά τους περιορισμούς αυτούς, το βιβλίο θα σας παρέχει τα απαραίτητα εργαλεία για να συνεχίσετε να μαθαίνετε αυτόνομα.
